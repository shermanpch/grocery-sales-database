```python
import os
from pathlib import Path


# Navigate to the project root directory (where .git is)
def find_git_root(path="."):
    path = Path(path).resolve()
    for parent in [path] + list(path.parents):
        if (parent / ".git").exists():
            return parent
    return path  # fallback if .git not found


project_root = find_git_root()
os.chdir(project_root)
print(f"Working directory set to: {project_root}")
```

```python
import inspect

import pandas as pd

import src.features
from src.feasy.sparkle import Sparkle

pd.set_option("display.max_columns", 50)
```

```python
data = pd.read_parquet("data/processed/customer_indexed.parquet")
records = list(data.itertuples(index=False))
data.head()
```

```python
feature_functions = []
# Iterate through members of the features module
for name, member in inspect.getmembers(src.features):
    if (
        inspect.isfunction(member)
        and member.__module__ == src.features.__name__
        and hasattr(member, "schema")
    ):
        feature_functions.append(member)

feature_functions
```

```python
feature_eng = Sparkle(feature_functions).source(records)
features = feature_eng.to_pandas()[0].set_index("CustomerID")
features.head()
```

```python
features.to_parquet("data/processed/features.parquet")
```
