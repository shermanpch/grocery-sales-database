```python
import os
from pathlib import Path


# Navigate to the project root directory (where .git is)
def find_git_root(path="."):
    path = Path(path).resolve()
    for parent in [path] + list(path.parents):
        if (parent / ".git").exists():
            return parent
    return path  # fallback if .git not found


project_root = find_git_root()
os.chdir(project_root)
print(f"Working directory set to: {project_root}")
```

```python
import pandas as pd
```

```python
sales = pd.read_csv("data/sales.csv")
sales.info()
sales.head()
```

```python
products = pd.read_csv("data/products.csv")
products.info()
products.head()
```

```python
categories = pd.read_csv("data/categories.csv")
categories.info()
categories.head()
```

```python
employees = pd.read_csv("data/employees.csv")
employees.info()
employees.head()
```

```python
countries = pd.read_csv("data/countries.csv")
countries.info()
countries.head()
```

```python
cities = pd.read_csv("data/cities.csv")
cities.info()
cities.head()
```

```python
customers = pd.read_csv("data/customers.csv")
customers.info()
customers.head()
```

```python
# Sales with ProductID -> CategoryID -> CategoryName
tmp = pd.merge(
    sales,
    products,
    on="ProductID",
    how="left",
    validate="m:1",
).merge(
    categories,
    on="CategoryID",
    how="left",
    validate="m:1",
)

tmp.head()
```

```python
# Sales with SalesPersonID (Employee)
employees = employees.rename(columns={"EmployeeID": "SalesPersonID"})
cols = [
    f"SalesPerson{col}" if col != "SalesPersonID" else col for col in employees.columns
]
employees.columns = cols

employee_cities = cities.copy()
cols = [f"SalesPerson{col}" for col in employee_cities.columns]
employee_cities.columns = cols

employee_countries = countries.copy()
cols = [f"SalesPerson{col}" for col in employee_countries.columns]
employee_countries.columns = cols

tmp = (
    pd.merge(
        tmp,
        employees,
        on="SalesPersonID",
        how="left",
        validate="m:1",
    )
    .merge(
        employee_cities,
        on="SalesPersonCityID",
        how="left",
        validate="m:1",
    )
    .merge(
        employee_countries,
        on="SalesPersonCountryID",
        how="left",
        validate="m:1",
    )
)

tmp.head()
```

```python
# Sales with CustomerID
cols = [f"Customer{col}" if col != "CustomerID" else col for col in customers.columns]
customers.columns = cols

customer_cities = cities.copy()
cols = [f"Customer{col}" for col in customer_cities.columns]
customer_cities.columns = cols

customer_countries = countries.copy()
cols = [f"Customer{col}" for col in customer_countries.columns]
customer_countries.columns = cols

tmp = (
    pd.merge(
        tmp,
        customers,
        on="CustomerID",
        how="left",
        validate="m:1",
    )
    .merge(
        customer_cities,
        on="CustomerCityID",
        how="left",
        validate="m:1",
    )
    .merge(
        customer_countries,
        on="CustomerCountryID",
        how="left",
        validate="m:1",
    )
)

tmp.head()
```
